```python
import numpy as np
```

```python
# sine wave as audio. 
rate = 48000  
x = np.arange(0, 5, 1/rate)
Audio(np.sin(x*2*np.pi*200), rate=rate, autoplay=True)
```

```python
import time
#  12 
ftable_full=np.logspace(0,np.log10(2),13)*262
#  do, re, me
ftable = ftable_full[[0,2,4,5,7,9,11]]
```

```python
# more sine waves
rate = 48000
x = np.arange(0, 3, 1/rate)
s1=np.sin(x*2*np.pi*ftable[0])*0.01
s2=np.sin(x*2*np.pi*ftable[2])*0.01
s3=np.sin(x*2*np.pi*ftable[4])*0.01

Audio(s1+s2+s3, rate=rate, autoplay=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot((s1+s2*0.5+s3)[:500])
```

```python
music = []
for s in [5,3,3,4,2,2,1,2,3,4,5,5,5]:
    x = np.arange(0, .5, 1/rate)
    display(Audio(np.sin(x*2*np.pi*ftable[s-1]), rate=rate, autoplay=True))
    time.sleep(0.5)
```

```python
# concatenate the tones together
music = []
for s in [5,3,3,4,2,2,1,2,3,4,5,5,5]:
    x = np.arange(0, .5, 1/rate)
    music.append(np.sin(x*2*np.pi*ftable[s-1]))
Audio(np.concatenate(music), rate=rate, autoplay=True)
```
